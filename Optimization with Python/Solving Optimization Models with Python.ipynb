# Question 1

```python
#QUESTION 1

model= ConcreteModel()

# declare variables
model.x1 = Var(domain=NonNegativeReals)
model.x2 = Var(domain=NonNegativeReals)

# objective function
model.Z = Objective(expr = 2*model.x1 + model.x2, sense=maximize)

#constraints
model.one = Constraint(expr = model.x2 <= 10)
model.two = Constraint(expr = 2*model.x1 + 5*model.x2 <=60)
model.three = Constraint(expr = model.x1 + model.x2 <=18)
model.four = Constraint(expr = 3*model.x1 + model.x2 <=44)

#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()

# display solutions
print('\nZ = ', model.Z())

print('\nx1 = ' , model.x1())
print('x2 = ' , model.x2())
```

# Question 2

```python
#QUESTION 2

model= ConcreteModel()

# declare variables
model.x1 = Var(domain=NonNegativeReals)
model.x2 = Var(domain=NonNegativeReals)
model.x3 = Var(domain=NonNegativeReals)
model.x4 = Var(domain=NonNegativeReals)
model.x5 = Var(domain=NonNegativeReals)

# objective function
model.mincost = Objective(expr = 22*model.x1 + 20*model.x2+ 25*model.x3+ 24*model.x4+ 27*model.x5, sense=minimize)

#constraints
model.tin = Constraint(expr = 60*model.x1 + 25*model.x2+ 45*model.x3+ 20*model.x4+ 50*model.x5 == 40)
model.zinc = Constraint(expr = 10*model.x1 + 15*model.x2+ 45*model.x3+ 50*model.x4+ 40*model.x5 == 35)
model.lead = Constraint(expr = 30*model.x1 + 60*model.x2+ 10*model.x3+ 30*model.x4+ 10*model.x5 == 25)

#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()

# display solutions
print('\nMinimumCost = ', model.mincost())

print('\nx1 = ' , model.x1())
print('x2 = ' , model.x2())
print('x3 = ' , model.x3())
print('x4 = ' , model.x4())
print('x5 = ' , model.x5())
```

# Question 3

```python
#QUESTION 3

model= ConcreteModel()

# declare variables
model.x1 = Var(bounds=(0,1000))
model.x2 = Var(bounds=(0,1200))

# objective function
model.profit = Objective(expr = 32*model.x1 + 24*model.x2, sense=maximize)

#constraints
model.nylon = Constraint(expr = 3*model.x1 + 2*model.x2 <= 5000)
model.labour = Constraint(expr = 45*model.x1 + 40*model.x2 <= 84000)


#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()

# display solutions
print('\nProfit = ', model.profit())

print('\nx1 = ' , model.x1())
print('x2 = ' , model.x2())
```

# Question 4

```python
#QUESTION 4

model= ConcreteModel()

# declare variables
model.x1 = Var(bounds=(0,2200))
model.x2 = Var(bounds=(0,3600))

# objective function
model.cost = Objective(expr = 1.2*model.x1 + 1.7*model.x2, sense=minimize)

#constraints
model.mingallon = Constraint(expr = model.x1 + model.x2 >= 3300)
model.maxgallon = Constraint(expr = model.x1 + model.x2 <= 4300)
model.octane = Constraint(expr = (88*model.x1 + 66*model.x2)>= 78*(model.x1+model.x2))


#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()

# display solutions
print('\nCost = ', model.cost())

print('\nx1 = ' , model.x1())
print('x2 = ' , model.x2())
```

```python
from pyomo.environ import *
```

```python
#QUESTION 1
# Create a model
model = ConcreteModel()

source = ['Plant1', 'Plant2', 'Plant3']

dest = ['DC1', 'DC2', 'DC3', 'DC4']

# declare variables
model.x = Var(source,dest, domain=NonNegativeReals)

# objective function
model.shippingcost = Objective(expr = 500*model.x[source[0], dest[0]] + 750*model.x[source[0], dest[1]] + 300*model.x[source[0], dest[2]] + 450*model.x[source[0], dest[3]] +
                         650*model.x[source[1], dest[0]] + 800*model.x[source[1], dest[1]] + 400*model.x[source[1], dest[2]] + 600*model.x[source[1], dest[3]] +
                         400*model.x[source[2], dest[0]] + 700*model.x[source[2], dest[1]] + 500*model.x[source[2], dest[2]] + 550*model.x[source[2], dest[3]], sense=minimize)


#constraints
model.plant1 = Constraint(expr = model.x[source[0], dest[0]] + model.x[source[0], dest[1]] + model.x[source[0], dest[2]] + model.x[source[0], dest[3]] == 12)
model.plant2 = Constraint(expr = model.x[source[1], dest[0]] + model.x[source[1], dest[1]] + model.x[source[1], dest[2]] + model.x[source[1], dest[3]] == 17)
model.plant3 = Constraint(expr = model.x[source[2], dest[0]] + model.x[source[2], dest[1]] + model.x[source[2], dest[2]] + model.x[source[2], dest[3]] == 11)

model.dc1 = Constraint(expr = model.x[source[0], dest[0]] + model.x[source[1], dest[0]] + model.x[source[2], dest[0]] == 10)
model.dc2 = Constraint(expr = model.x[source[0], dest[1]] + model.x[source[1], dest[1]] + model.x[source[2], dest[1]] == 10)
model.dc3 = Constraint(expr = model.x[source[0], dest[2]] + model.x[source[1], dest[2]] + model.x[source[2], dest[2]] == 10)
model.dc4 = Constraint(expr = model.x[source[0], dest[3]] + model.x[source[1], dest[3]] + model.x[source[2], dest[3]] == 10)



#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()


print ('\nShipment from source to destination: \n')

for s in source:
  for d in dest:
    print(s, 'to', d, ':', model.x[s, d](), 'trucks')

```

```python
#QUESTION 2
# Create a model
model = ConcreteModel()

source = ['PlantA', 'PlantB']

dest = ['DC1', 'DC2', 'DC3', 'Dummy']

# declare variables
model.x = Var(source,dest, domain=NonNegativeReals)

# objective function
model.shippingcost = Objective(expr = 800*model.x[source[0], dest[0]] + 700*model.x[source[0], dest[1]] + 400*model.x[source[0], dest[2]] + 0*model.x[source[0], dest[3]] +
                         600*model.x[source[1], dest[0]] + 800*model.x[source[1], dest[1]] + 500*model.x[source[1], dest[2]] + 0*model.x[source[1], dest[3]], sense=minimize)


#constraints
model.plantA = Constraint(expr = model.x[source[0], dest[0]] + model.x[source[0], dest[1]] + model.x[source[0], dest[2]] + model.x[source[0], dest[3]] == 50)
model.plantB = Constraint(expr = model.x[source[1], dest[0]] + model.x[source[1], dest[1]] + model.x[source[1], dest[2]] + model.x[source[1], dest[3]] == 50)


model.dc1 = Constraint(expr = model.x[source[0], dest[0]] + model.x[source[1], dest[0]] == 20)
model.dc2 = Constraint(expr = model.x[source[0], dest[1]] + model.x[source[1], dest[1]] == 20)
model.dc3 = Constraint(expr = model.x[source[0], dest[2]] + model.x[source[1], dest[2]] == 20)
model.dc4 = Constraint(expr = model.x[source[0], dest[3]] + model.x[source[1], dest[3]] == 40)



#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()


print ('\nShipment from source to destination: \n')

for s in source:
  for d in dest:
    print(s, 'to', d, ':', model.x[s, d](), 'trucks')

```
